## Consulta de datos en Kaggle

```python
import kaggle
import pandas as pd

# Competencia específica
competencia = 'dm-ey-f-2024-tercera'

# Obtener todos los submissions
submissions = kaggle.api.competitions_submissions_list(competencia)

# Inicializar listas para almacenar los datos
envios = []
cortes = []
public_scores = []
# private_scores = []

# Limitar a los últimos 100 submissions
num_submissions = 500  # Puedes cambiar esto a 150 o el número que necesites
submissions_limited = submissions[:num_submissions]

# Iterar sobre los submissions y extraer los datos requeridos
for submission in submissions_limited:
    # Obtener el nombre completo del envío (antes del último "_")
    file_name = submission['fileName']
    envio = '_'.join(file_name.split('_')[:-1])  # Tomamos todo antes del último "_"
    
    # Obtener el corte (lo que está después del último "_" y antes de ".csv")
    corte = file_name.split('_')[-1].replace('.csv', '')
    
    # Obtener el valor público
    public_score = submission['publicScore']
    # private_scores = submission['privateScore']
    
    # Añadir los valores a las listas
    envios.append(envio)
    cortes.append(corte)
    public_scores.append(public_score)
    # private_scores.append(private_scores)

# Crear un DataFrame con los datos
df_submissions = pd.DataFrame({
    'Envio': envios,
    'Corte': cortes,
    'ValorPublico': public_scores
    # 'ValorPrivado': private_scores
})

# Mostrar las primeras filas del DataFrame
print(df_submissions.head())

# Guardar el DataFrame en un archivo CSV si lo necesitas
df_submissions.to_csv('submissions_kaggle.csv', index=False)

df_submissions
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Asegurarse de que el DataFrame está limpio
# Por ejemplo, eliminar filas con valores nulos en la columna 'PublicScore' (si es necesario)
df_submissions = df_submissions.dropna(subset=['ValorPublico'])

# Pivotear el DataFrame para obtener los valores correctos
pivot_df = df_submissions.pivot(index='Corte', columns='Envio', values='ValorPublico')

# Asegurarnos de que los datos en el DataFrame están en el tipo correcto
pivot_df = pivot_df.apply(pd.to_numeric, errors='coerce')

# Calcular el promedio de todas las columnas
promedio_linea = pivot_df.mean(axis=1)

# Graficar los datos
plt.figure(figsize=(12, 8))

# Graficar cada columna de 'Envio' como una línea
for envio in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[envio], marker='o', label=envio)

# Graficar la línea del promedio
plt.plot(pivot_df.index, promedio_linea, color='black', linestyle='--', label='Promedio General', linewidth=2)

# Etiquetas y título
plt.title('Valor Público en función de los Cortes para cada Envío')
plt.xlabel('Corte')
plt.ylabel('Valor Público')
plt.legend(title='Envíos', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()

# Mostrar el gráfico
plt.show()
```
